## Merge all 12 months data into a single file

```python

files = [file for file in os.listdir('./Sales_Data')]

li = []
    
for file in files:
    df = pd.read_csv('./Sales_Data/' + file, index_col=None, header=0)
#    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

all_months_df = pd.concat(li, axis=0, ignore_index=True)

all_months_df.to_csv('Sales_2019', index=False)
```

```python
all_months_df = pd.read_csv('Sales_2019')
all_months_df
```

## Q1. What was the best month for sales ? How much was earned that month ? 

### best if we convert the order_date from string to date as we can do more with dates - need to drop na

```python
q1_df = all_months_df[all_months_df['Order Date'].notna()]
```

### After removing the NA, there is still error in conversion so we use errors = 'coerce', then drop the bad ones

```python
q1_df['OrderDateTime'] = pd.to_datetime(q1_df['Order Date'], format='%m/%d/%y %H:%M', errors='coerce')
#xx = pd.to_datetime(all_months_df['Order Date'].iloc[0], format='%m/%d/%y %H:%M')
#xx
#xx.month
#all_months_df['month'] = all_months_df['Order Date'].dt.month
#all_months_df.info()
```

## the other bad records are the column headers from the files (NaT) - remove them

```python
q1_df.loc[q1_df['OrderDateTime'].isnull()].index
#q1_df.drop(q1_df['OrderDateTime'].isnull().index, inplace=True)
```

```python
q1_df['Sales'] = q1_df['Price Each'] * q1_df['Quantity Ordered']
```

## Q4. What products are most often sold together ?

```python
q4_df = pd.read_csv('q3_df')     # start from where we left
q4_df['OrderDateTime'] = pd.to_datetime(q4_df['OrderDateTime'])
q4_df
```

```python
# concatenate-strings-from-several-rows-using-pandas-groupby
q4_select = q4_df[['Order ID','Product']]
q4_select['Product'] = q4_select.groupby(['Order ID'])['Product'].transform(lambda x: ','.join(x))

#q4_df['Product'] = q4_df[['Order ID','Product']].groupby(['Order ID'])['Product'].transform(lambda x: ','.join(x))
#q1_new = q1_df.groupby(['Order ID'])['Product'].transform(lambda x: ','.join(x))
```

```python
q4_select
```

```python
q4_select = q4_select.drop_duplicates()  
q4_select
```

```python
q4_select['Product']
```

```python
# need to get only records sold in combination so look for comma
filt = q4_select['Product'].str.contains(',', na=False)
q4_select = q4_select.loc[filt]
q4_select
```

```python
combo_count = q4_select.groupby('Product')['Product'].count()
```

```python
combo_count.sort_values()
```

```python
# Another way to identify duplicates
q4_dups = q4_df[q4_df['Order ID'].duplicated(keep=False)]
q4_dups
```

```python
# Refer to stackoverflow counting-unique-pairs-of-numbers-into-a_python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in q4_select['Product']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
```

```python
#count
for key, value in count.most_common(10):
    print(key, value)
```

```python
# transform-a-counter-object-into-a-pandas-dataframe
df_combo = pd.DataFrame.from_dict(count, orient='index').reset_index()
```

```python
df_combo
```

```python
df_combo.rename(columns={'index': 'Product Combo', 0: 'Count'}, inplace=True)
```

```python
df_combo.sort_values(by='Count', ascending=False)
```

## What product sold the most ?

```python
q5_df = pd.read_csv('q3_df')     # start from where we left in q4 not q4
```

```python
q5_df
```

```python
product_count = q5_df.groupby('Product')['Quantity Ordered'].sum()
```

```python
product_count.sort_values(ascending=False)
```

```python
product_count
```

```python
plt.bar(product_count.index, product_count.values)
plt.ylabel('Qty Ordered')
plt.xlabel('Product')
plt.xticks(product_count.index, rotation='vertical', size=8)
plt.show()
```

```python
prices = q5_df.groupby('Product')['Price Each'].mean()
prices
```

```python
# add a second y-axis to matplotlib

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(product_count.index, product_count.values, color='g')
ax2.plot(product_count.index, prices.values, 'b-')

ax1.set_xlabel('Product')
ax1.set_xticklabels(product_count.index, rotation='vertical', size=8)
ax1.set_ylabel('Qty Ordered', color='g')
ax2.set_ylabel('Price', color='b')

plt.show()
```
