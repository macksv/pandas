## extract the median salary for programming languages

```python
age_filt = (df.Age > 17) & (df.Age < 71)
df_age_sal_prog = df.loc[age_filt, ['Age','LanguageWorkedWith', 'ConvertedComp']]
df_age_sal_prog.rename(columns={'ConvertedComp': 'SalaryUSD'}, inplace=True)
df_age_sal_prog.dropna(inplace=True)

df_by_age = df_age_sal_prog.groupby(['Age'])

df_by_age_all = df_by_age.median()
df_by_age_all.rename(columns={'SalaryUSD': 'All (SalaryUSD)'}, inplace=True)
df_by_age_all
```

### Notice that the age is not exact, we need to truncate to int

```python
df_age_sal_prog['Age'] = df_age_sal_prog['Age'].apply(lambda param: int(param))
```

```python
df_by_age = df_age_sal_prog.groupby(['Age'])
df_by_age_all = df_by_age.median()
df_by_age_all.rename(columns={'SalaryUSD': 'All (SalaryUSD)'}, inplace=True)

df_by_age_all.head(10)
```

```python
python_filt = df_age_sal_prog['LanguageWorkedWith'].str.contains('Python', na=False)
df_age_sal_python = df_age_sal_prog.loc[python_filt, ['Age','LanguageWorkedWith', 'SalaryUSD']]
df_age_sal_python
```

```python
df_by_age = df_age_sal_python.groupby(['Age'])
df_by_age_python = df_by_age.median()
df_by_age_python.rename(columns={'SalaryUSD': 'Python(SalaryUSD)'}, inplace=True)
df_by_age_python.head(10)
```

```python
js_filt = df_age_sal_prog['LanguageWorkedWith'].str.contains('JavaScript', na=False)
df_age_sal_js = df_age_sal_prog.loc[js_filt, ['Age','LanguageWorkedWith', 'SalaryUSD']]
df_age_sal_js
```

```python
df_by_age = df_age_sal_js.groupby(['Age'])
df_by_age_js = df_by_age.median()
df_by_age_js.rename(columns={'SalaryUSD': 'JS(SalaryUSD)'}, inplace=True)
df_by_age_js.head(10)
```

```python
df_age_prog_sal = pd.concat([df_by_age_all, df_by_age_python, df_by_age_js], axis=1)
df_age_prog_sal
df_age_prog_sal.to_csv('AgeProgSal.csv')
```

## Using 1. Apply - calling functions and applying to Series

```python
df.Country.value_counts()
```

```python
def convert_upper(param):
    if param is None:    # NaN is a float
        param = "X"
    return str(param).upper()
```
